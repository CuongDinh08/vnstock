### Update a cell range with a DataFrame

```python
worksheet = gc.open(sheet_file).sheet1 # M·ªü sheet m·∫∑c ƒë·ªãnh ƒë·ªÉ xu·∫•t d·ªØ li·ªáu, ch·ªçn 1 trong 3 t√πy ch·ªçn v·ªõi bi·∫øn worksheet, d√πng d√≤ng n√†o b·ªè comment d√≤ng ƒë√≥, v√† comment d√≤ng kh√¥ng c·∫ßn d√πng

cell_list = worksheet.range('A1:C2') # Ch·ªçn v√πng d·ªØ li·ªáu tr√™n Google Sheets b·∫°n mu·ªën ch√®n d·ªØ li·ªáu t·ª´ DataFrame
# Update the data in the worksheet

for cell, value in zip(cell_list, df.values.flatten()):
    cell.value = value

worksheet.update_cells(cell_list)
```
